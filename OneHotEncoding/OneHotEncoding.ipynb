```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Assuming your data is in a variable called 'data' and you want to one-hot encode the first column
column_transformer = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [0])], remainder='passthrough')
X2 = column_transformer.fit_transform(X2)
```

```python
X2
```

```python
m3 = LinearRegression()
```

```python
m3.fit(X2, y2)
```

```python
m3.predict([[0,2,1,	69000	,18000]])
```

```python
import pickle
```

```python
with open("ONEHOTENCODING", "wb") as f:
  pickle.dump(m3, f)
```

```python
with open("ONEHOTENCODING", "rb") as f:
  model1 = pickle.load(f)
```

```python
model1.predict([[0,2,1, 23000, 2311]])
```

```python
from joblib import dump, load
with open("ONEHOTENCODING", "wb") as f:
  dump(m3, f)
```
